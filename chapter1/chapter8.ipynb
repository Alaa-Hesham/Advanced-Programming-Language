```python
import numpy as np

arr = np.arange(1, 11)

mean = arr.mean()
median = np.median(arr)
std = arr.std()

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)
```

```python
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "Mona", "Omar"],
    "Age": [20, 22, 21, 23],
    "Score": [75, 88, 92, 60]
}

df = pd.DataFrame(data)

filtered = df[df["Score"] > 80]
print(filtered)
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Square Numbers Plot")
plt.show()
```

```python
import torch

a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

dot = torch.dot(a, b)
elementwise = a * b

print("Dot Product:", dot)
print("Element-wise Multiplication:", elementwise)
```
