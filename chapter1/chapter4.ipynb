```python
import re
m = re.match(r"(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})", "2023-05-12")
print(m.group("year"))  
```

```python
import re
text = "Python   is   great"
normalized = re.sub(r"\s+", " ", text)
print(normalized)  
```

```python
import re

emails = ["user@example.com", "bad-email", "test@domain.org"]
pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.(com|org|edu)$"

for email in emails:
    if re.match(pattern, email):
        print(f"Valid: {email}")
    else:
        print(f"Invalid: {email}")
```

```python
import re
text = "I love #Python and #AI"
hashtags = re.findall(r"#\w+", text)
print(hashtags)  
```

```python
import re
pattern = r"(\+\d-\d{3}-\d{4})|(\d{3}-\d{3}-\d{4})"

numbers = ["+1-555-1234", "123-456-7890", "5551234"]
for num in numbers:
    if re.match(pattern, num):
        print(f"Valid: {num}")
    else:
        print(f"Invalid: {num}")
```

```python
import re
text = "Python, Python! AI is great; Python AI."
words = re.findall(r"\b\w+\b", text)
freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1
print(freq)
```

```python
import re
text = "This is is a test test"
pattern = r'\b(\w+)\s+\1\b'
duplicates = re.findall(pattern, text)
full_duplicates = [f"{word} {word}" for word in duplicates]

print(full_duplicates)
```

```python
import re
text = "The events are on 2023-05-12 and 2024-01-01."
pattern = r'\b\d{4}-\d{2}-\d{2}\b'
dates = re.findall(pattern, text)

print(dates)
```

```python
import re
text = "Card: 1234-5678-9012-3456"
masked = re.sub(r"\d(?=(?:\D*\d){4})", "*", text)
print(masked)
```

```python
import re
text = "I know Python, Java, and C++ but not Ruby."
languages = re.findall(r"(?<!\w)(Python|Java|C\+\+|Ruby)(?!\w)", text)
print(languages)
```
