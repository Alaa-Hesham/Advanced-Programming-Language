```python
products = [" LAPTOP ", "phone ", " Tablet", "CAMERA "]
cleaned = list(map(lambda p: p.strip().title(), products))
print(cleaned)
```

```python
celsius = [0, 10, 20, 30, 40]
fahrenheit = list(map(lambda c: (9/5)*c + 32, celsius))
print(fahrenheit)
```

```python
nums = [1, 2, 3, 4, 5]
result = list(map(lambda x: x**2 + 10, nums))
print(result)
```

```python
words = ["python", "lambda", "programming", "map", "function"]
chars = list(map(lambda w: (w[0], w[-1]), words))
print(chars)
```

```python
marks = [[45, 80, 70], [90, 60, 100], [88, 76, 92]]
updated_marks = list(map(lambda row: list(map(lambda x: round(x * 1.05), row)), marks))
print(updated_marks)
```

```python
numbers = [10, 20, 30, 40, 50]
min_num = min(numbers)
max_num = max(numbers)

normalized = list(map(lambda x: (x - min_num) / (max_num - min_num), numbers))
print(normalized)
```

```python
sentences = ["Python is fun", "Map and lambda", "Advanced programming"]
lengths = list(map(lambda s: list(map(lambda w: len(w), s.split())), sentences))
print(lengths)
```
