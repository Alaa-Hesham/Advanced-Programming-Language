```python
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return "".join(ch for ch in text if ch not in vowels)
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

print(fib(10))  
```

```python
def fibonacci(n):
    if n<=1:
        return n
        return fibonacci(n-1)+fibonacci(n-2)
```

```python
def make_adder(n):
    def adder(x):
        return n + x
    return adder

add5 = make_adder(5)
```

```python
def apply_twice(func, value):
    return func(func(value))

result = apply_twice(lambda x: x + 1, 5)
```

```python
from collections import Counter

def etl_pipeline(texts):
    stopwords = {"the", "a", "is", "and", "an", "to"}

    tokens = [word.lower() for line in texts for word in line.split()]

    filtered = list(filter(lambda w: w not in stopwords, tokens))

    return dict(Counter(filtered))

data = ["The cat is here", "A cat and a dog"]
```

```python
def my_reduce(func, iterable, initial=None):
    iterator = iter(iterable)

    if initial is None:
        value = next(iterator)
    else:
        value = initial

    for item in iterator:
        value = func(value, item)

    return value
```

```python
def log_call(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}...")
        result = func(*args, **kwargs)
        print(f"{func.__name__} finished.")
        return result
    return wrapper

@log_call
def say_hello():
    print("Hello!")
```
